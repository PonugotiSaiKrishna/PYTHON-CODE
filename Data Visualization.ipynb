```python
import openpyxl   #openpyxl will create a excel file
import pandas as pd

workbook = openpyxl.Workbook()  
sheet = workbook.active

data = [
    ['NAME', 'DOMAIN', 'AGE', 'LOCATION', 'SALARY', 'EXP'],
    ['ALEX', 'TESTING', 25, 'BNG', 5000, 2],
    ['BARB', 'JAVA', 30, 'CHE', 10000, 3],
    ['CHERRY', 'C', 35, 'PUNE', 15000, 4],
    ['DIPAN', 'DA', 38, 'MUMBAI', 20000, 5],
    ['ESWAR', 'DS', 40, 'HYD', 50000, 6]
]

for row in data:
    sheet.append(row)

workbook.save('data.xlsx')
```

```python
import numpy as np #numpy library for matrix (Non-Dimentional Arrays)
import matplotlib.pyplot as plt #matplotlibs for visualizations
import seaborn as sns #seaborn library is for more complex statistical visualizations
```

```python
vis1 = sns.displot(emp['SALARY']) #displot will give the output as ploting the data in histogram form
```

```python
vis2 = sns.distplot(emp['SALARY']) #distplot will give the output bu ploting the data in distributed format
```

```python
vis3 = plt.hist(emp['SALARY']) #Using matplotlib as plt created a histogrom with hist
```

```python
plt.rcParams['figure.figsize'] = 5,1 
#rc plot :  is a dictionary-like object where you can set various parameters for customizing the appearance and behavior of your plots.
#figure.figsize :  specifically controls the default dimensions (in inches) of the figure when you create a plot.
```

```python
vis5 = sns.lmplot(data=emp, x= 'EXP', y='SALARY') 
#linear model plot : lmplot
#Seaborn is used to create a linear regression plot
```

```python
vis7 = sns.lmplot(data=emp, x= 'EXP', y='SALARY', fit_reg = False) 
#fit_reg=False disables the fitting of a regression line. This means you will only see the scatter plot of the data without the linear regression line.
```
